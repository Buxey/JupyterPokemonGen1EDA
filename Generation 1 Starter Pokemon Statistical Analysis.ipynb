```python
# We are trying to figure out which starter 
# pokemon from generation 1 statistically is the best option to choose and why
# this does not include gym leader battles
# or their evolved forms
# we are building our analysis just on the pokemon when they are selected from professor Oak
```

```python
# numpy the most popular Python library for array manipulation and numeric computing
import numpy as np
# matplotlib the most popular visualization library in the Python ecosystem.
import matplotlib.pyplot as plt
# Pandas is mainly used for data analysis
import pandas as pd
```

```python
# importing the dataset
df = pd.read_csv("/Users/Apple/Google Drive/Personal Programming/Python Exploratory Data Analysis/Pokemon Starter EDA/Pokemon.csv")
```

```python
# this is allowing use to take a glimps at our data
df.head()
```

```python
# filtering our first generation pokemon

bulbasaur = df[df.Name == 'Bulbasaur']
charmander = df[df.Name == 'Charmander']
squirtle = df[df.Name == 'Squirtle']
```

```python
pokemon = ('Bulbasaur (318)', 'Charmander (309)','Squirtle (314)')
y_pos = np.arange(len(pokemon))
total_value = [bulbasaur.Total.sum(), charmander.Total.sum(),squirtle.Total.sum()]

barlist = plt.bar(y_pos, total_value, align='center', alpha=1)
barlist[0].set_color('green')
barlist[1].set_color('red')
barlist[2].set_color('lightblue')
plt.xticks(y_pos, pokemon)
plt.ylabel('Total')
plt.title('Pokemon Total Comparison')

plt.show()

# First goes to Bulbasaur
# Second goes to Squirtle
# Third goes to Charmander
```

```python
# I have created a scoring system for the pokemon
# the scores have been taken from each of the graphs, 
# the pokemon that scored the highest for that graph will be given
# the highest number of points and so on

# first place is 3 points
# second place is 2 points
# third place is 1 point



pokemon = ('Bulbasaur (18)', 'Charmander (12)','Squirtle (12)')
y_pos = np.arange(len(pokemon))
attack_value = [18,12,12]

barlist = plt.bar(y_pos, attack_value, align='center', alpha=1)
barlist[0].set_color('green')
barlist[1].set_color('red')
barlist[2].set_color('lightblue')
plt.xticks(y_pos, pokemon)
plt.ylabel('Total')
plt.title('Pokemon Scores')

plt.show()

# the pokemon that scored the highest in each individual graph with a total of 18 is bulbasaur
# statistically showing that the best starter pokemon to choose in generation 1 is Bulbasaur
```
